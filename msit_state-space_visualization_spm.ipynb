# Visualize State-Space Estimates

```python
import os
import numpy as np
import pandas as pd
import warnings
from IPython.display import display
import xlwt
import xlrd
import csv
import sys
import matplotlib.pyplot as plt
import os, subprocess
import pandas as pd
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
from IPython.display import display
from scipy import stats

warnings.filterwarnings('ignore')

width=0.5

df_subjid = pd.read_csv('behavior_excel/subjid_to_analyze.txt', delimiter=',', dtype=str, names = ["subjid"])
df_case_control = pd.read_csv('behavior_excel/case_control.csv', names=['ID'])

df_subjid = df_subjid['subjid']
df_case_control = df_case_control['ID']

csfont = {'fontname': 'Verdana'}

cases = []
controls=[]

i=0
j=0

for subject, case in zip(df_subjid, df_case_control):
    if case == 1:
        cases.append(subject)
    elif case == 2:  
        controls.append(subject)

#~~~~~~~~~~~~~~~~~~~~~~~~~~~#
## Make Giant Case Database
#~~~~~~~~~~~~~~~~~~~~~~~~~~#

case_base = pd.DataFrame()
control_base = pd.DataFrame()

for case in cases:
    df_case=pd.read_csv('state_space/stan_results/%s_estimates.csv'%case)
    case_merge = [case_base, df_case]
    case_base = pd.concat(case_merge, axis=1)

for control in controls:
    df_control = pd.read_csv('state_space/stan_results/%s_estimates.csv'%control)
    control_merge = [control_base,df_control]
    control_base = pd.concat(control_merge, axis=1)

reversals = 3
n_presentations=17
subplot = 131
fig = plt.figure(figsize = (50,20))
plt.suptitle('State Space Model Averages: Cases v Controls', fontsize=65)
for j in range(0,reversals):
    
    plt.subplot(subplot)
    case_avg = np.zeros(n_presentations)
    control_avg = np.zeros(n_presentations)
    
    for i in range(0,n_presentations):
        case_avg[i]=np.mean(case_base.ix[(17*j)+i])
        control_avg[i]=np.mean(control_base.ix[(17*j)+i])
    
    
    plt.bar(np.array(range(1,18))-width, control_avg, color='skyblue', width=width, label='Control')
    plt.bar(np.array(range(1,18)),case_avg, color='m',width=width, label='Case')
    
    if j==0:
        plt.xlabel('Learning Phase', fontsize=55, **csfont)
    if j==1:
        plt.xlabel('Reversal 1', fontsize=55, **csfont)
    if j==2:
        plt.xlabel('Reversal 2', fontsize=55, **csfont)
    plt.tick_params(axis='x', which='major', labelsize=40)
    plt.tick_params(axis='y', which='major', labelsize=0.5, color='w')
    plt.xticks(np.arange(1, 18, 1.0))
    plt.gca().spines['left'].set_position(('data', -0.5))
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    if j > 0:
        plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_position(('data', 0))
    #plt.gca().spines['bottom'].set_position(('outward', 10))
    plt.gca().spines['right'].set_position(('data', 6))
    plt.gca().spines['bottom'].set_smart_bounds(True)
    if j==0:
        plt.tick_params(axis='y', which='major', labelsize=40)
        plt.legend(loc='upper left', fontsize=50)
        plt.ylabel('%Learned', fontsize=55, **csfont)
    plt.ylim(0,1)
    subplot+=1
plt.tight_layout()
plt.show()

fig.savefig('State_Space_CasevControl.jpg')
```
